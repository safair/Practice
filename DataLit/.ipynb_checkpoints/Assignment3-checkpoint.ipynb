# World Map (Time Series for GDP, GDP per Capita, Education and Homicide)

```python
def draw_map(dataset, target_var, title):
    

    years = list(map(str, range(1995,2017)))
    # make list of continents
    #continents = dataset.Region.unique().tolist()
    # make figure
    figure = {
        'data': [],
        'layout': {},
        'frames': []
    }

    # fill in most of layou
    figure['layout']['title'] = title
    figure['layout']['geo'] = dict(showframe = False,
                               showcoastlines = False,
                               projection = dict(
                                   type = 'mercator'
                               ))
    figure['layout']['hovermode'] = 'closest'
    figure['layout']['sliders'] = {
        'args': [
            'transition', {
                'duration': 400,
                'easing': 'cubic-in-out'
            }
        ],
        'initialValue': '1995',
        'plotlycommand': 'animate',
        'values': years,
        'visible': True
    }
    figure['layout']['updatemenus'] = [
        {
            'buttons': [
                {
                    'args': [None, {'frame': {'duration': 500, 'redraw': False},
                             'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                    'label': 'Play',
                    'method': 'animate'
                },
                {
                    'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                    'transition': {'duration': 0}}],
                    'label': 'Pause',
                    'method': 'animate'
                }
            ],
            'direction': 'left',
            'pad': {'r': 10, 't': 87},
            'showactive': False,
            'type': 'buttons',
            'x': 0.1,
            'xanchor': 'right',
            'y': 0,
            'yanchor': 'top'
        }
    ]

    sliders_dict = {
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 20},
            'prefix': 'Year:',
            'visible': True,
            'xanchor': 'right'
        },
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'pad': {'b': 10, 't': 0},
        'len': 0.9,
        'x': 0.1,
        'y': 0,
        'steps': []
    }

    # make data
    year = 2016
  
    figure['data'].append(map_dict(dataset, target_var, year))

    # make frames
    for year in years:
        frame = {'data': [], 'name': str(year)}
        
       
        data_dict = map_dict(data, target_var, int(year))
        frame['data'].append(data_dict)

        figure['frames'].append(frame)
        slider_step = {'args': [
            [year],
            {'frame': {'duration': 300, 'redraw': False},
             'mode': 'immediate',
           'transition': {'duration': 300}}
         ],
         'label': year,
         'method': 'animate'}
        sliders_dict['steps'].append(slider_step)


    figure['layout']['sliders'] = [sliders_dict]
    iplot(figure)

    
    #div = plotly.offline.plot( figure,filename='d3-world-map')

    return figure
```

```python
def map_dict(data, target_var, year=1995):
    data =  dict(
        type = 'choropleth',
        locations = data['Code'],
        z = data.loc[data.Year==year][target_var],
        text = data['Country'],
        colorscale = [[0,"rgb(165, 0, 38)"],[0.35,"rgb(215, 48, 39)"],[0.5,"rgb(244, 109, 67)"],\
            [0.6,"rgb(253, 174, 97)"],[0.9,"rgb(254, 224, 144)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        
        colorbar = dict(
            title = f"{target_var} {year}",
        )
      ) 
    
    return data
```

```python
div = draw_map(data, 'GDP', 'GDP')
```

```python
target = ['Population','GDP', 'GDP_per_capita', 'Expected_years_of_schooling', 'Homicide']
```

```python
for t in target:
    draw_map(data, t, t)
```

# Comparing Different Features

```python
def draw_compare(dataset, x,x_title, y, y_title, weight, range_val):
    sz = (dataset[weight].describe()['50%'] - dataset[weight].describe()['25%'])//20  
    dataset = dataset.fillna(0)
    
    years = list(map(str, range(1995,2017)))
    # make list of continents
    continents = dataset.Region.unique().tolist()
    # make figure
    figure = {
        'data': [],
        'layout': {},
        'frames': []
    }

    # fill in most of layou
    figure['layout']['xaxis'] = {'range': [dataset[x].min(), dataset[x].max()], 'title': x_title}
    figure['layout']['yaxis'] = {'range': range_val['num'], \
                                 'type':range_val['type'], 'title': y_title}
    figure['layout']['hovermode'] = 'closest'
    figure['layout']['sliders'] = {
        'args': [
            'transition', {
                'duration': 400,
                'easing': 'cubic-in-out'
            }
        ],
        'initialValue': '1995',
        'plotlycommand': 'animate',
        'values': years,
        'visible': True
    }
    figure['layout']['updatemenus'] = [
        {
            'buttons': [
                {
                    'args': [None, {'frame': {'duration': 500, 'redraw': False},
                             'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                    'label': 'Play',
                    'method': 'animate'
                },
                {
                    'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                    'transition': {'duration': 0}}],
                    'label': 'Pause',
                    'method': 'animate'
                }
            ],
            'direction': 'left',
            'pad': {'r': 10, 't': 87},
            'showactive': False,
            'type': 'buttons',
            'x': 0.1,
            'xanchor': 'right',
            'y': 0,
            'yanchor': 'top'
        }
    ]

    sliders_dict = {
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 20},
            'prefix': 'Year:',
            'visible': True,
            'xanchor': 'right'
        },
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'pad': {'b': 10, 't': 50},
        'len': 0.9,
        'x': 0.1,
        'y': 0,
        'steps': []
    }

    # make data
    year = 2000
    for continent in continents:
        dataset_by_year = dataset[dataset['Year'] == year]
        dataset_by_year_and_cont = dataset_by_year[dataset_by_year['Region'] == continent]

        data_dict = {
            'x': list(dataset_by_year_and_cont[x]),
            'y': list(dataset_by_year_and_cont[y]),
            'mode': 'markers',
            'text': list(dataset_by_year_and_cont['Country']),
            'marker': {
                'sizemode': 'area',
                'sizeref': sz,
                'size': list(dataset_by_year_and_cont[weight])
            },
            'name': continent
        }
        figure['data'].append(data_dict)

    # make frames
    for year in years:
        frame = {'data': [], 'name': str(year)}
        for continent in continents:
            dataset_by_year = dataset[dataset['Year'] == int(year)]
            dataset_by_year_and_cont = dataset_by_year[dataset_by_year['Region'] == continent]

            data_dict = {
                'x': list(dataset_by_year_and_cont[x]),
                'y': list(dataset_by_year_and_cont[y]),
                'mode': 'markers',
                'text': list(dataset_by_year_and_cont['Country']),
                'marker': {
                    'sizemode': 'area',
                    'sizeref': sz,
                    'size': list(dataset_by_year_and_cont[weight])
                },
                'name': continent
            }
            frame['data'].append(data_dict)

        figure['frames'].append(frame)
        slider_step = {'args': [
            [year],
            {'frame': {'duration': 300, 'redraw': False},
             'mode': 'immediate',
           'transition': {'duration': 300}}
         ],
         'label': year,
         'method': 'animate'}
        sliders_dict['steps'].append(slider_step)


    figure['layout']['sliders'] = [sliders_dict]
    return figure
```

```python
#draw_compare(dataset, x,x_title, y, y_title, weight)
range_val = {'num':[0,10], 'type':'log'}
fig = draw_compare(data, 'Expected_years_of_schooling', 'Education','GDP_per_capita', 'GDP per Capita', 'Population', range_val)
iplot(fig)
```

```python
#draw_compare(dataset, x,x_title, y, y_title, weight)
range_val = {'num':[0,5], 'type':'linear'}
fig = draw_compare(data, 'Expected_years_of_schooling', 'Education','Homicide', 'Homicide', 'Population', range_val)
iplot(fig)
```

```python
#draw_compare(dataset, x,x_title, y, y_title, weight)
range_val = {'num':[0,5], 'type':'linear'}
fig = draw_compare(data, 'GDP_per_capita', 'GDP_per_capita','Homicide', 'Homicide', 'Population', range_val)
iplot(fig)
```
